```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
# Generate 200 random samples with 2 features
np.random.seed(42)  # For reproducibility
x = np.random.randn(200, 2) * 2  # 200 samples, 2 features with some scaling
# Generate corresponding binary labels
y = np.where(x[:, 0] + x[:, 1] > 0, 1, 0)  # Simple decision boundary for binary classification

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
x_test
```

```python
train=LogisticRegression()
train.fit(x_train,y_train)
```

```python
y_pre=train.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pre)
```
